## An exerise to generate random variables and visualize their distributions:
### 1. Normal 
### 2. Bernoulli 
### 3. Binomial 
### 4. Student's t 

What are random variables, you ask? They are basically functions of numbers that give output between 0 and 1. They are widely used in describing population, for example, the distribution of GRE scores, given enough sample size, can be represented by a Normal distribution with a certain mean and standard deviation. Note that a Normal R.V. is a continuous random variable, this means that the input space is all of the real line (as opposed to just integers). The parameters associated with a Normal distribution are its mean $\mu$ and variance $\sigma^2$

Bernoulli random variables, on the other hand, are discrete variables that are often used to describe something that has a 'binary' behavior. For example, tossing a fair coin can be represented by a Bernoulli R.V. with p = 0.5. The parameter associated with a Bernoulli R.V. is the probability of success of one binary trial, usually denoted as just $p$

Binomial random variables are similar to Bernoulli, except that it describes the success of many more binary trials. The parameters for a Binomial R.V. are $n$ and $p$, where $n$ is the number of trials and $p$ is the probability of success

Student's t distributions are commonly used in statistical testing, for example, describing the significance of difference between two sample means in a hypothesis testing experiment. It's parameters are characterized by it's degress of freedom $\nu$ (pronounced nu). Notice that the higher $\nu$ gets, the closer the t distribution is to the standard normal distrbution ($\mu$ = 0 and $\sigma^2$ = 1) 

```python
# First import useful packages 
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

import math
%matplotlib inline
```

```python
# import the normal module from scipy
from scipy.stats import norm
# define parameters for Normal distribution 
# mu = mean
mu = 0 
# s2 = variance 
s2 = 1 
# s = standard deviation 
s = np.sqrt(s2)
# define the domain:
X1 = np.linspace(start=-10,stop=10,num=1000)
# define the probability density function 
Y1= norm.pdf(X1,mu,s)
plt.plot(X1,Y1)
```

```python
# what if we vary the mean?
plt.figure(1,figsize=(10,10))
for m0 in np.linspace(-3,3,5):
    mu = m0
    s2=1
    s=np.sqrt(s2)
    Y1= norm.pdf(X1,mu,s)
    plt.plot(X1,Y1)
```

```python
# what if we vary the variance?
plt.figure(2,figsize=(10,10))
# Note : variance and standard deviation are magnitudes of data spread and are therefore always positive
mu=0
for s0 in np.linspace(1,10,10):
    s2= s0
    s=np.sqrt(s2)
    Y1=norm.pdf(X1,mu,s)
    plt.plot(X1,Y1)
```

```python
# what about varying both? 
plt.figure(3,figsize=(10,10))
for m0 in np.linspace(-3,3,5):
    for s0 in np.linspace(1,10,10):
        mu=m0
        s=np.sqrt(s0)
        Y1=norm.pdf(X1,mu,s)
        plt.plot(X1,Y1)
```
